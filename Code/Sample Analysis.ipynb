```
df2 = pd.DataFrame(columns=['uri','domain','suffix'])
```

```
#count of mysql mentions
for record in f22:
    uri2 = record.header.get('warc-target-uri', None)
    if not uri2:
        continue
    #domain = urlparse(uri).netloc
    text2 = record.payload.read()
    if 'mysql'.encode('utf-8') in text2:
        sfx2 = tldextract.extract(uri2).suffix
        domain2 = tldextract.extract(uri2).domain
        row2 = {'uri':uri2, 'domain':domain2, 'suffix':sfx2}
        df2 = df2.append(row2, ignore_index=True)
```

```
#count of database mentions
for record in f22:
    uri2 = record.header.get('warc-target-uri', None)
    if not uri2:
        continue
    #domain = urlparse(uri).netloc
    text2 = record.payload.read()
    if 'database'.encode('utf-8') in text2:
        sfx2 = tldextract.extract(uri2).suffix
        domain2 = tldextract.extract(uri2).domain
        row2 = {'uri':uri2, 'domain':domain2, 'suffix':sfx2}
        df2 = df2.append(row2, ignore_index=True)
```
